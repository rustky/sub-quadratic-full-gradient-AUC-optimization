```python
import pandas as pd
```

```python
header = ['epoch', 'train loss','train auc','test loss', 'test auc', 'lr']
libauc_dt = pd.read_csv('LibAUC Results/ResNet20/STL10-0.5-0.316227766016838-100', names=header,delimiter=' ', index_col=False)
libauc_dt
```

```python
functional_dt = pd.read_csv('results-2022-01-17/pretrained=True-loss_name=square_test-lr=0.316227766016838-batch_size=100-imratio=0.5-dataset=STL10-model=ResNet20.csv', sep = '\t')
functional_dt
```

```python
import matplotlib.pyplot as plt
plt.plot(libauc_dt['epoch'],libauc_dt['test auc'],'r-',
        functional_dt['epoch'],functional_dt['test_auc'],'b-')
plt.ylabel("AUC")
plt.xlabel("Epochs")
plt.legend(["LibAUC Test AUC", "Hocking Rust Test AUC"],
          bbox_to_anchor=(1, 0.9))
plt.title("Libauc vs Us")
plt.show()
```
